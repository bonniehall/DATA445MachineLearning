```python
## Q1 Using the pandas library, read the csv data file and create a data-frame called house_price.

import pandas as pd
import boto3
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import tensorflow as tf


## Defining the S3 bucket
s3 = boto3.resource('s3')
bucket_name = 'bonnieh-data-445-bucket'
bucket = s3.Bucket(bucket_name)

## Defining the file to be read from s3 bucket
file_key = 'CarPrice_Assignment.csv'

bucket_object = bucket.Object(file_key)
file_object = bucket_object.get()
file_content_stream = file_object.get('Body')

##R Reading the csv file
car_price = pd.read_csv(file_content_stream)
car_price.head()
```

```python
## Q2  Using the wheelbase, enginesize, compressionratio, horsepower, peakrpm, citympg, as
## the predictor variables, and price is the target variable, split the data into train (80%) and test (20%)

## defining the input variable and price
X=car_price[['wheelbase', 'enginesize', 'compressionratio', 'horsepower', 'peakrpm', 'citympg']]
Y = car_price['price']

## split the data

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)
```

```python
## Q3 transforming to 0-1 scale
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
## Q4 Using the train dataset, build a MLP with a single hidden layer with 10 neurons and the
## ReLU as the activation function. Also use optimizer = â€™adamâ€™, epochs = 100 and batch size =
## 100. After that, use this model to predict on the test dataset. 
## Report the MSE of this model.

md1 = tf.keras.models.Sequential([
    tf.keras.layers.Dense(10, input_dim = 6, activation = 'relu'),
    tf.keras.layers.Dense(1)
])

md1.compile(optimizer = 'adam', loss = 'mse')
md1.fit(X_train, Y_train, epochs = 100, batch_size = 100)

## md1.fit(X_train, Y_train, epochs = 100, batch_size = 100, verbose = 0)
```

```python
## making predictions on the test set

md1.evaluate(X_test, Y_test)
```

```python
## to see the weights

md1.layers[0].weights
```

```python
## Q5 Using the train dataset, build a MLP with two hidden layers: the first one with 10 neurons
## and the ReLU as the activation function, the second one with 8 neurons and ReLU as the activation
## function. Also use optimizer = â€™adamâ€™, epochs = 100 and batch size = 100. After that, use this
## model to predict on the test dataset. Report the MSE of this model.

md2 = tf.keras.models.Sequential([
    tf.keras.layers.Dense(10, input_dim = 6, activation = 'relu'),
    tf.keras.layers.Dense(10, activation = 'relu'),
    tf.keras.layers.Dense(1)
])

md2.compile(optimizer = 'adam', loss = 'mse')
md2.fit(X_train, Y_train, epochs = 100, batch_size = 100, verbose = 0) 
```

```python
## predicting on the test data

md2.evaluate(X_test, Y_test)
```
