```python
## Q1 Load the data file to your S3 bucket. Using the pandas library, read the csv data file
## and create a data-frame called car price.

import pandas as pd
import boto3
import numpy as np
## import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
## for regression is RandomForestRegressor
from sklearn.linear_model import LogisticRegression, Lasso, LassoCV
from sklearn.metrics import recall_score


## Defining the S3 bucket
s3 = boto3.resource('s3')
bucket_name = 'bonnieh-data-445-bucket'
bucket = s3.Bucket(bucket_name)

## Defining the file to be read from s3 bucket
file_key = 'framingham.csv'

bucket_object = bucket.Object(file_key)
file_object = bucket_object.get()
file_content_stream = file_object.get('Body')

##R Reading the csv file
heart = pd.read_csv(file_content_stream)
heart.head()
```

```python
## Q2 Remove observations with missing values.

heart = heart.dropna()
heart.head()
```

```python
## Q3 Using male, age, currentSmoker, totChol, sysBP,BMI, heartRate, and glucose as the
## predictor variables, and TenYearCHD is the target variable, split the data into two data-frames (taking
## into account the proportion of 0s and 1s): train (80%) and test (20%).

X = heart[['male', 'age', 'currentSmoker', 'totChol', 'sysBP', 'BMI', 'heartRate','glucose']]
Y = heart['TenYearCHD']
           
X_train, X_test, Y_train,Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y)        
X_train.head()
```

```python
## Q4  Transform each of the input variables in the train and test dataset to 0-1 scale.

scaler = MinMaxScaler()

X_train = pd.DataFrame(scaler.fit_transform(X_train), columns = X_train.columns)
X_test = pd.DataFrame(scaler.fit_transform(X_test), columns = X_test.columns)
X_test.head()
```

```python
## Q5  Using the train data-frame, build a LASSO model in which male, age, currentSmoker,
## totChol, sysBP,BMI, heartRate, and glucose are the predictor variables, and TenYearCHD is the target
## variable. After that, using the top 5 variables from the LASSO model (the variables with the highest
## estimated coefficients in absolute value), build a logistic regression model in which TenYearCHD is the
## target variable. Using this model, predict the likelihood of risk of coronary disease of the patients in
## the test data-frame. Using 10% as cutoff value, report the recall.

## LASSO

lasso_cv = LassoCV(cv = 5).fit(X_train, Y_train)
lasso_cv.alpha_

## Building LASSO regression
lasso_md = Lasso(alpha = lasso_cv.alpha_).fit(X_train, Y_train)
lasso_md.coef_

## lists the coefficients for each variable
## order of the variables: 'male', 'age', 'currentSmoker', 'totChol', 'sysBP', 'BMI', 'heartRate','glucose'
```

```python
## Keeping only the 5 variables with the best LASSO scores

X_train_lasso = X_train[['male', 'age', 'currentSmoker', 'sysBP', 'glucose']]
X_test_lasso = X_test[['male', 'age', 'currentSmoker', 'sysBP', 'glucose']]

## Logistic Regression
logit_md = LogisticRegression().fit(X_train_lasso, Y_train)

## predicting on test
pred1 = logit_md.predict_proba(X_test_lasso)[:,1]
pred1 = np.where(pred1 < 0.1, 0, 1)

## computing the recall
recall_score(Y_test, pred1)
```

```python
## Q6 Using the train data-frame, build a random forest model (with 500 trees) in which: male,
## age, currentSmoker, totChol, sysBP,BMI, heartRate, and glucose are the predictor variables, and
## TenYearCHD is the target variable. After that, using the top 5 variables from the random forest model
## (the variables with the highest importance values), build a random forest model in which TenYearCHD
## is the target variable. Using this model, predict the risk of coronary disease of the patients in the test
## data-frame. Using 10% as cutoff value, report the recall.

## building RF classifier

RF = RandomForestClassifier(n_estimators = 500).fit(X_train, Y_train)

## extracting importance
RF.feature_importances_

## want highest values--which are 5, 6, 4, 8, 2
## Variable list: 'male', 'age', 'currentSmoker', 'totChol', 'sysBP', 'BMI', 'heartRate','glucose'
```

```python
## creating the test data with only the 5 best variables

X_train_RF = X_train[['age','totChol','sysBP', 'BMI', 'glucose']]
X_test_RF = X_test[['age','totChol','sysBP', 'BMI', 'glucose']]
X_train_RF.head()
```

```python
## creating the RF model

RF = RandomForestClassifier(n_estimators = 500, max_depth = 5).fit(X_train_RF, Y_train)

## predicint on test
pred2 = RF.predict_proba(X_test_RF)[:,1]
pred2 = np.where(pred2 < 0.1, 0, 1)

## calculating the recall
recall_score(Y_test, pred2)
```
