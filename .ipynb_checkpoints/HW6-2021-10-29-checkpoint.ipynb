```python
## Q2 Explain why AbaBoost.M1 is an ensemble learning algorithm? Be specific.

## Ensemble learning means that predictions from multiple models are used for the final prediction. With
## AdaBoost.M1 the weights are adjusted with each round and new predictions are made. The final prediction is made
## using a weighted average of all of the weak learners. In this way, a strong learner is essentially the outcome.
```

```python
## Q6a.  Load the Framingham.csv data file to you S3 bucket. Using the pandas library, read the csv data
## file and create a data-frame called heart. Remove observation with missing values.

import pandas as pd
import numpy as np
import boto3
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier, RandomForestClassifier
from sklearn.metrics import recall_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
import matplotlib.pyplot as plt

## Defining the s3 bucket
s3 = boto3.resource('s3')
bucket_name = 'bonnieh-data-445-bucket'
bucket = s3.Bucket(bucket_name)

## Defining the file to be used
file_key = 'framingham.csv'

bucket_object = bucket.Object(file_key)
file_object = bucket_object.get()
file_content_stream = file_object.get('Body')

## Reading the csv file
heart = pd.read_csv(file_content_stream)

## Drop any NA values
heart = heart.dropna()
heart
```

Q6b Based on the different interactions that we have so far with this data file, it
seems that age, totChol, sysBP,BMI, heartRate, and glucose are the import variables for
predicting the likelihood of TenYearCHD. Using these six variables as the predictor variables
and TenYearCHD as the target variable, do the following:

(i) Split the data into train (80%) and test (20%) (taking into account the proportion of
0s and 1s).

(ii) Transform all the input variables from the train and test dataset to 0-1 scale.

(ii) Using the train data-frame, build a random forest classifier (using 500 trees and maximum depth tree equal to 3). Using this model, predict the likelihood of risk of coronary
disease of the patients in the test data-frame. Using 10% as cutoff value, report the
recall.

(iii) Using the train data-frame, build an AdaBoost classifier (using 500 trees and maximum
depth tree equal to 3). Using this model, predict the likelihood of risk of coronary
disease of the patients in the test data-frame. Using 10% as cutoff value, report the
recall.

(iv) Using the train data-frame, build an AdaBoost classifier (use 50 learners and the support vector machine classifier as the learner with rbf as the kernel. Donât forget to
include probability = True in the support vector machine model; otherwise, you will
run into errors.). Using this model, predict the likelihood of risk of coronary disease of
the patients in the test data-frame. Using 10% as cutoff value, report the recall.

Repeat (i)-(iv) 100 times. Compute the average recall of each of the models across the 100
iterations. What model would use to predict TenYearCHD? Be specific.

```python
## Defining the variables

X= heart[['totChol', 'sysBP', 'BMI', 'heartRate', 'glucose']]
Y = heart['TenYearCHD']

## Defining lists to store recall scores
recalls_rf = []
recalls_ada = []
recalls_svm = []

for i in range(0, 100):
    ## Splitting the data
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size =0.2, stratify = Y)

    ## Building and fitting the Random Forest model
    rf_md = RandomForestClassifier(n_estimators = 500, max_depth = 3).fit(X_train, Y_train)

    ## predicting test data with Random Forest model
    pred_rf = rf_md.predict_proba(X_test)[:,1]
    pred_rf = np.where(pred_rf < 0.1, 0, 1)

    ## calculating Random Forest recall score 
    recall_rf = recall_score(Y_test, pred_rf)
    recalls_rf.append(recall_rf)
    
    ## Building and fitting the AdaBoost Classifier
    ada_md = AdaBoostClassifier(base_estimator = DecisionTreeClassifier(max_depth = 3), n_estimators = 500).fit(X_train, Y_train)

    ## predicting test data with AdaBoost Classifier
    pred_ada = ada_md.predict_proba(X_test)[:,1]
    pred_ada = np.where(pred_ada < 0.1, 0, 1)

    ## calculating the AdaBoost Classifier recall score
    recall_ada = recall_score(Y_test, pred_ada)
    recalls_ada.append(recall_ada)
    
    ## Building and fitting the AdaBoost SVM Classifier
    svm_md = AdaBoostClassifier(base_estimator = SVC(kernel = 'rbf',  probability = True), n_estimators = 50).fit(X_train, Y_train)

    ## Predicting test data using AdaBoost SVM Classifier
    pred_svm = svm_md.predict_proba(X_test)[:,1]
    pred_svm = np.where(pred_svm < 0.1, 0, 1)

    ## Calculating AdaBoose SVM recall score
    recall_svm = recall_score(Y_test, pred_svm)
    recalls_svm.append(recall_svm)
```

```python
## plotting the recall values

plt.figure(figsize=(15, 5))
plt.plot(recalls_rf, label = 'Recall Random Forest')
plt.plot(recalls_ada, label = 'Recall AdaBoost Classifier')
plt.plot(recalls_svm, label = 'Recall AdaBoost SVM')
plt.ylabel('Recall score')
plt.xlabel('iteration number')
plt.legend(loc = 'lower right')
```

```python
## calculating the average recall values
print('The average recall for the Random Forest model is', round(np.mean(recalls_rf), 4), 
      ', the average recall for the AdaBoost classifier model is', round(np.mean(recalls_ada), 4),
      ', and the average recall for the AdaBoost SVM model is',round(np.mean(recalls_svm), 4))
```

```python
## Based on both the graph and the average recall values, I would pick the AdaBoost classifier model. Both The AdaBoost Classifier and the 
## AdaBoost SVM had 100% recall. But the SVM took substantially longer for one run, which makes it more computationally expensive.
## Also, decision trees are easier to explain than an SVM model.
```
