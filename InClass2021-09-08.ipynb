```python
## Q1 Load the data file to you S3 bucket. Using the pandas library, read the csv data file and
## create a data-frame called car_price.

import pandas as pd
import boto3

## Defining the S3 bucket
s3 = boto3.resource('s3')
bucket_name = 'bonnieh-data-445-bucket'
bucket = s3.Bucket(bucket_name)

## Defining the file to be read from s3 bucket
file_key = 'CarPrice_Assignment.csv'

bucket_object = bucket.Object(file_key)
file_object = bucket_object.get()
file_content_stream = file_object.get('Body')

## Reading the csv file
car_price = pd.read_csv(file_content_stream)
car_price
```

```python
## Q2 Using wheelbase, horsepower, peakrpm, citympg, and highwaympg as the predictor variables, and price 
## is the target variable build a linear regression model. Perform a leave-one-out cross
## validation. Report the MSE of this model.

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import LeaveOneOut
import numpy as np

## Defining input and target variables

X = car_price[['wheelbase', 'horsepower', 'peakrpm', 'citympg', 'highwaympg']]
Y = car_price['price']

## Defining the list to store the results
md1_results = list()

## Defining the cross validation
cv = LeaveOneOut()

for train_ix, test_ix in cv.split(X):
    
    ## splitting the data
    X_train, X_test = X.loc[train_ix], X.loc[test_ix]
    Y_train, Y_test = Y.loc[train_ix], Y.loc[test_ix]
    
    ## Building the linear model
    lm_md = LinearRegression().fit(X_train, Y_train)
    
    ## Predicting on the test set
    pred = lm_md.predict(X_test)
    
    ## Storing the results
    md1_results.append(pred[0])
    
## Computing the MSE
mse_md1 = np.mean(np.power(md1_results-Y, 2))
mse_md1
```

```python
## Q3 Using carwidth, carheight, horsepower, citympg, and highwaympg as the predictor variables, and price 
## is the target variable build a linear regression model. Perform a leave-one-out cross
## validation. Report the MSE of this model.

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import LeaveOneOut

## Defining input and target variables

X = car_price[['carwidth', 'carheight', 'horsepower', 'citympg', 'highwaympg']]
Y = car_price['price']

## Defining the list to store the results
md2_results = list()

## Defining the cross validation
cv = LeaveOneOut()

for train_ix, test_ix in cv.split(X):
    
    ## splitting the data
    X_train, X_test = X.loc[train_ix], X.loc[test_ix]
    Y_train, Y_test = Y.loc[train_ix], Y.loc[test_ix]
    
    ## Building the linear model
    lm_md = LinearRegression().fit(X_train, Y_train)
    
    ## Predicting on the test set
    pred = lm_md.predict(X_test)
    
    ## Storing the results
    md2_results.append(pred[0])
    
## Computing the MSE
mse_md2 = np.mean(np.power(md2_results-Y, 2))
mse_md2
```
